```python
excelname= "kd"
sheetname = "kd"

wb = openpyxl.load_workbook(str(excelname)+".xlsx")
ws = wb[str(sheetname)]

dataset = []
i = 1
while(ws.cell(row=i, column=1).value!=None):
    row = []
    j = 1
    while(ws.cell(row=i, column=j).value!=None):
        row.append(ws.cell(row=i, column=j).value)
        j = j +1
    dataset.append(row)
    i = i + 1
```

```python
pd.DataFrame(dataset)
```

```python
trade = []
for i in range(0, len(dataset)):
    row = []
    for j in range(0, 11):
        row.append(0)
    trade.append(row)
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "rsv"
trade[0][6] = "k"
trade[0][7] = "d"
trade[0][8] = "buy_day"
trade[0][9] = "cum_buy_day"
trade[0][10] = "sell_day"
```

```python
#æŠŠkdæª”æ¡ˆä¸­çš„æ•¸æ“šæŠ“é€²ä¾†
for i in range(1, len(trade)):
    trade[i][0] = dataset[i][0]
    trade[i][1] = float(dataset[i][1])
    trade[i][2] = float(dataset[i][2])
    trade[i][3] = float(dataset[i][3])
    trade[i][4] = float(dataset[i][4])


#RSV(t) = (close(t) - min(low(z)| z = t, t-1, ..., t-8)/(max(high(z)) - min(low(z)| z = t, t-1, ..., t-8)*100 , t = 9, 11, ..., T
for i in range(9, len(trade)):
    rsv = 0
    high = []
    low = []
    for j in range(0, 9):
        high.append(trade[i-j][2])
        low.append(trade[i-j][3])
    rsv = (trade[i][4]-min(low))/(max(high)-min(low))*100
    trade[i][5] = rsv
    print(rsv)
```

```python
#k (Kå€¼ åˆå§‹å€¼ = 50, at day 9ï¼Œæ‰€ä»¥ K(9) = 50)
trade[9][6] = 50
#K(t)= K(t-1)*2/3 + RSV(t)*1/3, t = 10, 11, ..., T
for i in range(10,len(trade)):
    K = 0 #æ¯æ¬¡é‡æ–°æ­¸é›¶ï¼Œé‡æ–°è¨ˆç®—Kå€¼
    K = (trade[i-1][6]*2/3) + (trade[i][5]*1/3)
    trade[i][6] = K
    print(K)
```

```python
#d (Då€¼ åˆå§‹å€¼ = 50, at day 9ï¼Œæ‰€ä»¥ D(9) = 50)
trade[9][7] = 50
#D(t)= D(t-1)*2/3 + K(t)*1/3 , t = 10, 11, ..., T
for i in range(10,len(trade)):
    D = 0 #æ¯æ¬¡é‡æ–°æ­¸é›¶é‡æ–°è¨ˆç®—Då€¼
    D =(trade[i-1][7]*2/3) + (trade[i][6]*1/3) #trade[i][6]ä¸èƒ½ç›´æŽ¥å¯«æˆKï¼Œå› ç‚ºå¯«KæœƒåªæŠ“åˆ°æœ€å¾Œä¸€ç­†Kå€¼
    trade[i][7] = D
    print(D)
```

```python
#KDé»ƒé‡‘äº¤å‰-å»ºè­°è²·é€²
#ç•¶KDæŒ‡æ¨™çš„Kå€¼ç”±ä¸‹å¾€ä¸Šçªç ´Då€¼ï¼Œå»ºè­°è²·é€²ã€åšå¤š
#K(t) > D(t)
for i in range(10, len(trade)):
    if(trade[i][6] > trade[i][7])and(trade[i-1][9]==0):
        trade[i][8] = 1
        trade[i][9] = 1
        trade[i][10] = 0
    elif(trade[i][6] < trade[i][7])and(trade[i-1][9]==1):
        trade[i][8] = 0
        trade[i][9] = 0
        trade[i][10] = 1
    elif(trade[i-1][2]>0):
        trade[i][8] = 0
        trade[i][9] = 1
        trade[i][10] = 0
```

```python
for i in range(0, len(trade)):
    for j in range(0, len(trade[1])):
        ws.cell(row=i+1, column=j+1).value= trade[i][j]
wb.save('kd3.xlsx')
```
